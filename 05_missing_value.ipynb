### 2-1 결측 데이터 필터링

### 연습문제 3)
1. 남자승객의 결측치는 남자 승객 평균으로 채우기 => 남자, 여자 승객 각각의 컨디션 만들기
2. 여자승객의 결측치는 여자 승객 평균으로 채우기
3. 결측치를 모두 채우고 나서 모든 사람의 평균 출력

```python
df_copy = df.copy()
```

```python
# 남자 승객의 나이 평균

cond1 = df_copy['sex'] == 'male'
```

```python
male_mean = df_copy.loc[cond1, 'age'].mean()

print(male_mean)
```

```python
# 여자 승객의 나이 평균

cond2 = df_copy['sex'] == 'female'
```

```python
female_mean = df_copy.loc[cond2, 'age'].mean()

print(female_mean)
```

```python
# 구한 평균으로 결측치 채우기 
### 컨디션이라는 변수에 접근해야 하므로 .loc

df_copy.loc[cond1, 'age'].fillna(male_mean)
df_copy.loc[cond2, 'age'].fillna(female_mean)
```

```python
# 결측치로 채운값을 원본 데이터에 저장하기 

df_copy.loc[cond1, 'age'] = df_copy.loc[cond1, 'age'].fillna(male_mean)
df_copy.loc[cond2, 'age'] = df_copy.loc[cond2, 'age'].fillna(female_mean)
```

```python
df_copy['age'].mean()
```
