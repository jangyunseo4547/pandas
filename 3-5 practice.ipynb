```
species (종): 펭귄의 종 (예: Adelie, Gentoo, Chinstrap)

bill_length_mm (부리 길이, mm): 펭귄의 부리 길이를 밀리미터(mm) 단위로 측정한 값
bill_depth_mm (부리 깊이, mm): 펭귄의 부리 높이(깊이)를 밀리미터(mm) 단위로 측정한 값
flipper_length_mm (날개 길이, mm): 펭귄의 날개(오리발) 길이를 밀리미터(mm) 단위로 측정한 값
body_mass_g (체중, g): 펭귄의 몸무게를 그램(g) 단위로 측정한 값
sex (성별): 펭귄의 성별 (Male 또는 Female)
```

### 3. Female Chinstrap종의 bill_length_mm 높은순 정렬

````python
df_copy['species'].value_counts()
````

````python
# 1) 컨디션 1 = chinstrap 종 / 컨디션 2 = 여자
cond1 = df_copy['species'] == 'Chinstrap'
cond2 = df_copy['sex'] == 'Female'

# 2) 둘 다 충족을 변수에 저장 
female_chinstrap = df_copy.loc[(cond1) & (cond2)] # 조건을 ()로 묶어줘야 함.

# 3) 내림차순으로 정렬 : 기준.sort_values([컬럼])
female_chinstrap_sorted = female_chinstrap.sort_values(['bill_length_mm'], ascending = False)

female_chinstrap_sorted.head(10)
````

### 4. species(종)별 성별 구분하여 평균 몸무게 출력
=> ex) Adelie, Female / Adelie, male / Gentoo, Female...

````python
df_copy['species'].value_counts()
````

````python
df_copy['sex'].value_counts()
````

````python
# 1) 종별 성별을 각각의 컨디션에 저장

cond1 = (df_copy['species'] == 'Adelie') & (df_copy['sex'] == 'Male')
cond2 = (df_copy['species'] == 'Adelie') & (df_copy['sex'] == 'Female')

cond3 = (df_copy['species'] == 'Gentoo') & (df_copy['sex'] == 'Male')
cond4 = (df_copy['species'] == 'Gentoo') & (df_copy['sex'] == 'Female')

cond5 = (df_copy['species'] == 'Chinstrap') & (df_copy['sex'] == 'Male')
cond6 = (df_copy['species'] == 'Chinstrap') & (df_copy['sex'] == 'Female')

# 2) 각 평균값 출력 
cond1_mean = df_copy.loc[cond1, 'body_mass_g'].mean()
cond2_mean = df_copy.loc[cond2, 'body_mass_g'].mean()
cond3_mean = df_copy.loc[cond3, 'body_mass_g'].mean()
cond4_mean = df_copy.loc[cond4, 'body_mass_g'].mean()
cond5_mean = df_copy.loc[cond5, 'body_mass_g'].mean()
cond6_mean = df_copy.loc[cond6, 'body_mass_g'].mean()

print(cond1_mean,cond2_mean,cond1_mean,cond3_mean,cond4_mean,cond5_mean,cond6_mean)
````
