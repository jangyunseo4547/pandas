## 8. where() 함수
- condition이 False인 조건에만 바꿈.
    - ex) 30보다 큰 경우 

```python
df.tail(5)
```

```python
cond = df['fare'] < 30 
df['fare'].where(cond, 1).tail(5) #30보다 작은 사람의 fare를 1로 바꾼다.
```

- 2) 그 중에서 tip을 가장 많이 낸 순서로 내림차순 정렬

```python
## 1) 정보 확인

tips.value_counts()
```

```python
## 1) 남성 흡연자 먼저 조건 필터링 (원본을 바꾸고 싶으면 변수에 저장해야 함.)

cond1 = tips['sex'] == 'Male'

cond2 = tips['smoker'] == 'Yes'

tips.loc[cond1 & cond2].sort_values('tip', ascending = False).head(10)
```

- 식사 인원이 5명 이상인 저녁테이블
- 흡연자가 없는 주말 테이블
- total_bil,tip 출력 

```python
## 2) 
cond1 = tips['size'] >= 5
cond2 = tips['time'] == 'Dinner'

cond3 = tips['smoker'] == 'No'
cond4 = tips['day'].isin(['Sat','Sun'])

# tips.loc[(cond1 & cond2) & (cond3 & cond4), ['total_bill', 'tip']]
tips.loc[(cond1 & cond2) & (cond3 & cond4)]
```
