```python
import seaborn as sns
import pandas as pd

df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
# 1) 성별을 기준으로 생존률의 평균 (남자 생존률 평균 / 여자 생존률 평균 보기)

df_copy = df.copy()
```

```python
# groupby : ~을 기준으로 그룹으로 묶어줌
df_copy.groupby('sex').mean(numeric_only = True)
```

```python
# 리스트 안에 컬럼으로 여러개를 기준으로 그룹바이 가능
df_copy.groupby(['sex', 'pclass']).mean(numeric_only = True)
```

```python
# 컬럼을 지정해서 그룹바이 하고 싶은 경우
df_copy.groupby(['sex', 'pclass'])[['survived']].mean(numeric_only = True)

# 대괄호 2개 : 데이터 프레임
# 대괄호 1개 : 시리즈
```

### reset_index()
- 인덱스를 다시 쪼개줌.

```python
df_temp = df_copy.groupby(['sex', 'pclass'])[['survived']].mean()
df_temp
```

```python
df_temp.reset_index()
```

```python
df_copy.groupby(['sex', 'pclass'])[['survived', 'age']].agg(['mean','sum'])
```

### pivot_table
- df_copy.pivot_table(index = 'who', values = 'survived') : 인덱스와 컬럼 지정

```python
# 한개의 인덱스

df_copy.pivot_table(index = 'who', values = 'survived')
```

```python
# 여러개의 인덱스

df_copy.pivot_table(index = ['who', 'pclass'], values = 'survived')
```
